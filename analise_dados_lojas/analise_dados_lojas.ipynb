# 2. Análise do faturamento



```python
faturamento_loja = sum(loja.preço) - sum(loja.frete)
faturamento_loja2 = sum(loja2.preço) - sum(loja2.frete)
faturamento_loja3 = sum(loja3.preço) - sum(loja3.frete)
faturamento_loja4 = sum(loja4.preço) - sum(loja4.frete)

print(f"Faturamento da loja 1: {faturamento_loja:.2f}\nFaturamento da loja 2: {faturamento_loja2:.2f}\nFaturamento da loja 3: {faturamento_loja3:.2f}\nFaturamento da loja 4: {faturamento_loja4:.2f}\n")
```

```python
count_tipos_pagamentos_loja = loja.tipo_de_pagamento.value_counts()
count_tipos_pagamentos_loja2 = loja2.tipo_de_pagamento.value_counts()
count_tipos_pagamentos_loja3 = loja3.tipo_de_pagamento.value_counts()
count_tipos_pagamentos_loja4 = loja4.tipo_de_pagamento.value_counts()
```

```python
count_tipos_pagamentos_loja
```

# Qual loja fechar?

Para responder a essa pergunta, é importante olhar por outro viés: **quais lojas devemos manter?**

## Loja 1 – Essencial para o faturamento
A **Loja 1** foi a que apresentou o maior ganho em faturamento e representa um ponto comercial essencial para a venda de eletrônicos e eletrodomésticos de maior valor agregado.  
Isso garante uma vantagem significativa em relação às outras lojas, tornando sua manutenção estratégica.

Entretanto, ao escolher manter essa loja, precisamos pensar em como a empresa deseja se posicionar.  
Nesta análise, consideramos que a empresa busca **diversificação de capital**.

Nesse sentido, percebe-se que a Loja 1 apresenta vendas nas categorias:
- Brinquedos  
- Esporte e lazer  
- Livros  
- Utilidades domésticas  

Porém, essas categorias possuem um baixo valor de venda por esta loja.

## Loja 2 – Complemento estratégico
Com base nas lojas que apresentam maior faturamento, a **Loja 2** surge como uma excelente segunda escolha para manter.  

Ela apresenta faturamento elevado justamente nas categorias em que a Loja 1 tem menor valor agregado por produto.  
Embora ambas vendam quantidades semelhantes, a Loja 2 pratica **preços médios mais altos**, o que fortalece o equilíbrio do portfólio da empresa.

## Loja 3 vs Loja 4 – A decisão final
Resta agora decidir entre **manter ou fechar a Loja 3 ou a Loja 4**.

O capital já se encontra bem diversificado, então o próximo critério relevante é:  
**as regiões estão sendo bem atendidas?**

Observa-se uma pequena vantagem da **Loja 3** em relação à Loja 4 em alguns estados, principalmente no **DF** e no **ES**, onde as Lojas 1 e 2 também possuem baixo alcance.

Além disso:
- A Loja 3 supera a Loja 4 em aproximadamente **R$ 70.000,00 em faturamento**  
- A Loja 3 apresenta **maior disponibilidade geográfica**

Esses fatores tornam a Loja 3 mais estratégica para manutenção.

## Conclusão – Loja a ser fechada
A **Loja 4** apresenta:
- Baixo faturamento  
- Pouca ou nenhuma vantagem geográfica  
- Baixa contribuição para a diversificação do capital  

Portanto, ela se mostra a melhor opção para ser fechada.

A avaliação das lojas é um critério importante para a tomada de decisão.
