# 2. Análise do faturamento



```python
faturamento_loja = sum(loja.preço) - sum(loja.frete)
faturamento_loja2 = sum(loja2.preço) - sum(loja2.frete)
faturamento_loja3 = sum(loja3.preço) - sum(loja3.frete)
faturamento_loja4 = sum(loja4.preço) - sum(loja4.frete)

print(f"Faturamento da loja 1: {faturamento_loja:.2f}\nFaturamento da loja 2: {faturamento_loja2:.2f}\nFaturamento da loja 3: {faturamento_loja3:.2f}\nFaturamento da loja 4: {faturamento_loja4:.2f}\n")
```

```python
count_tipos_pagamentos_loja = loja.tipo_de_pagamento.value_counts()
count_tipos_pagamentos_loja2 = loja2.tipo_de_pagamento.value_counts()
count_tipos_pagamentos_loja3 = loja3.tipo_de_pagamento.value_counts()
count_tipos_pagamentos_loja4 = loja4.tipo_de_pagamento.value_counts()
```

```python
count_tipos_pagamentos_loja
```

```python
faturamento_tipo_pagamento_loja = loja.preço.groupby(loja.tipo_de_pagamento).sum() - loja.frete.groupby(loja.tipo_de_pagamento).sum()
faturamento_tipo_pagamento_loja2 = loja2.preço.groupby(loja.tipo_de_pagamento).sum() - loja2.frete.groupby(loja.tipo_de_pagamento).sum()
faturamento_tipo_pagamento_loja3 = loja3.preço.groupby(loja.tipo_de_pagamento).sum() - loja3.frete.groupby(loja.tipo_de_pagamento).sum()
faturamento_tipo_pagamento_loja4 = loja4.preço.groupby(loja.tipo_de_pagamento).sum() - loja4.frete.groupby(loja.tipo_de_pagamento).sum()
```

```python
faturamento_tipo_pagamento_loja
```

A diferença entre as duas lojas com menor faturamento:

```python
dif_ultimos =  faturamento_loja3 - faturamento_loja4
dif_ultimos
```

# 8. Gráficos e visualização de dados

```python
df_barras_categoria_quantidade = pd.DataFrame({
    "Loja 1": count_categoria_produto_loja,
    "Loja 2": count_categoria_produto_loja2,
    "Loja 3": count_categoria_produto_loja3,
    "Loja 4": count_categoria_produto_loja4
})

df_barras_categoria_quantidade.plot(kind="bar")
plt.title("Quantidade de vendas por categoria (comparação entre lojas)")
plt.ylabel("Quantidade")
plt.xlabel("Categoria")
plt.xticks(rotation=45)
plt.show()
```

```python
df_barras_categoria_faturamento = pd.DataFrame({
    "Loja 1": faturamento_categoria_loja,
    "Loja 2": faturamento_categoria_loja2,
    "Loja 3": faturamento_categoria_loja3,
    "Loja 4": faturamento_categoria_loja4
})

df_barras_categoria_faturamento.plot(kind="bar")
plt.title("Faturamento por categoria (comparação entre lojas)")
plt.ylabel("Faturamento")
plt.xlabel("Categoria")
plt.xticks(rotation=45)
plt.show()
```

```python
df_barras_tipo_pagamento_faturamento = pd.DataFrame({
    "Loja 1": faturamento_tipo_pagamento_loja,
    "Loja 2": faturamento_tipo_pagamento_loja2,
    "Loja 3": faturamento_tipo_pagamento_loja3,
    "Loja 4": faturamento_tipo_pagamento_loja4
})

df_barras_tipo_pagamento_faturamento.plot(kind="bar")
plt.title("Faturamento por forma de pagamento (comparação entre lojas)")
plt.ylabel("Faturamento")
plt.xlabel("Forma de pagamento")
plt.xticks(rotation=45)
plt.show()
```

```python
df_barras_regiao_faturamento = pd.DataFrame({
    "Loja 1": faturamento_regiao_loja.sort_values(ascending = False).head(5),
    "Loja 2": faturamento_regiao_loja2.sort_values(ascending = False).head(5),
    "Loja 3": faturamento_regiao_loja3.sort_values(ascending = False).head(5),
    "Loja 4": faturamento_regiao_loja4.sort_values(ascending = False).head(5)
})

df_barras_regiao_faturamento.plot(kind="bar")
plt.title("Faturamento por região (comparação entre lojas)")
plt.ylabel("Faturamento")
plt.xlabel("Região")
plt.xticks(rotation=45)
plt.show()
```
